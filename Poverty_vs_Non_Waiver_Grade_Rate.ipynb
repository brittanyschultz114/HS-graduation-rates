```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
```

```python
poverty_rate_df = pd.read_csv("Resources/combined_poverty_rate_revised.csv")
school_grade_15_df = pd.read_csv("Resources/schoolgrade2015_revised.csv")
school_grade_16_df = pd.read_csv("Resources/schoolgrade2016_revised.csv")
school_grade_17_df = pd.read_csv("Resources/schoolgrade2017_revised.csv")
school_grade_18_df = pd.read_csv("Resources/schoolgrade2018_revised.csv")
school_grade_19_df = pd.read_csv("Resources/schoolgrade2019_revised.csv")
grad_rate_15_df = pd.read_csv("Resources/2015_grad_rates_revised.csv")
grad_rate_16_df = pd.read_csv("Resources/2016_grad_rates_revised.csv")
grad_rate_17_df = pd.read_csv("Resources/2017_grad_rates_revised.csv")
grad_rate_18_df = pd.read_csv("Resources/2018_grad_rates_revised.csv")
grad_rate_19_df = pd.read_csv("Resources/2019_grad_rates_revised.csv")
school_grade_16_df.head()
```

```python
grade_list_15 = school_grade_15_df['Corp Grade 2015']
num_list_15 = []
letter_grade = ['','A','B','C','D','F']
for grade in grade_list_15:
    if grade == 'A':
        num_list_15.append(1)
    elif grade == 'B':
        num_list_15.append(2)
    elif grade == 'C':
        num_list_15.append(3)
    elif grade == 'D':
        num_list_15.append(4)
    elif grade == 'F':
        num_list_15.append(5)
        
school_grade_15_df['Number'] = num_list_15
unique = np.unique(letter_grade, return_inverse=True)
grade_15 = school_grade_15_df['Number']
pov_15 = poverty_rate_df['2015 Poverty Rate']
grad_rate_15 = grad_rate_15_df['State Non-Waiver Grad Rate'].str.replace('%','').astype(np.float64)
# line regression for 2015 pov rate and grad rate.
(slope, intercept, rvalue, pvalue, stderr) = linregress(grad_rate_15,pov_15)
regress_values = grad_rate_15 * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.figure(figsize=(8,6))
plt.scatter(grad_rate_15,pov_15,c=grade_15)
plt.plot(grad_rate_15,regress_values,"r-")
plt.annotate(line_eq,(40,5),fontsize=15,color="red")
# plt.xticks(range(len(letter_grade)), letter_grade)
# plt.xlim(.5, 5.5)
clb = plt.colorbar(ticks=[1, 2, 3, 4, 5])
clb.set_ticklabels(['A','B','C','D','F'])
clb.set_label("District Grade")
plt.title("2015 Poverty Rate vs. Non-Waiver Grad Rate vs. District Grade")
plt.xlabel("Graduation Rate")
plt.ylabel("Poverty Rate")
plt.grid(alpha=.5)
plt.show

cc_grad_pov_2015 = round(st.pearsonr(grad_rate_15,pov_15)[0],2)
print(f"The correlation coefficient between non-waiver graduation rate and poverty rate is {cc_grad_pov_2015}")

plt.savefig('Output_Data/2015_PovRate_NonWaiverGradRate.png')
```

# School Scores A - F

This notebook holds code for cleaning the data on Indiana school letter grades obtained from the Indiana deparment of Education's webiste: https://www.doe.in.gov/accountability/find-school-and-corporation-data-reports

In looking for data on Indiana schools and their graduation rates, there were other scores of information, but the Indiana Department of education seemed to be the authority on the matter. We also wanted to evaluate the same data that our state leaders were using in their decision making process. Our goal was to gain clearity on the matter of graduation waivers given to students that did not meet Indiana high school graducation requirments in spite of attending four years of high school. The school letter grades are given to schools base on an array of factors. We want to see how school grades relate to the use of graduation waivers. 

```python
#Dependencies
import pandas as pd
```

```python
# Read csv files with school scores from state of IN 2015, 2016, 2017, 2018, and2019. 
school_grade2015 = pd.read_csv('Resources/A-F_IN_school_grades2015.csv')
school_grade2016 = pd.read_csv('Resources/A-F_IN_school_grades2016.csv')
school_grade2017 = pd.read_csv('Resources/A-F_IN_school_grades2017.csv')
school_grade2018 = pd.read_csv('Resources/A-F_IN_school_grades2018.csv')
school_grade2019 = pd.read_csv('Resources/A-F_IN_school_grades2019.csv')
#school_grade2015.head()
```

```python
# Check out data
school_grade2015.head()
```

```python
# Isolate only data needed
df2015 = school_grade2015[["Corp ID","Corp Name","Corp Grade 2015"]]
df2015.head()
```

```python
# Rename Corporate Name column to match other years
corp_grade2015 = df2015.rename(columns={"Corp Name":"Corporate Name"})
corp_grade2015.head()
```

```python
# Same process for other years: Check out data
school_grade2016.head()
```

```python
# Isolate data needed
df2016 = school_grade2016[["Corp ID", "Corp Name", "2016 Corp Grade"]]
df2016.head()
```

```python
# Rename columns to match
corp_grade2016 = df2016.rename(columns={"Corp Name":"Corporate Name",
                                       "2016 Corp Grade":"Corp Grade 2016"})
corp_grade2016.head()
```

```python
# Check it out
school_grade2017.head()
```

```python
# Isolate needed info
df2017 = school_grade2017[["Corp ID", "Corporation Name", "2016-17 Grade"]]
df2017.head()
```

```python
# Rename columns
corp_grade2017 = df2017.rename(columns={"2016-17 Grade":"Corp Grade 2017"})
corp_grade2017.head()
```

```python
# Check out data
school_grade2018.head()
```

```python
# Isolate needed data
df2018 = school_grade2018[["Corp", "Corp Name", "2017-2018 Grade"]]
df2018.head()
```

```python
# Rename columns to match other years
corp_grade2018 = df2018.rename(columns={"Corp Name":"Corporate Name",
                                       "2017-2018 Grade":"Corp Grade 2018"})
corp_grade2018.head()
```

```python
# Check out data
school_grade2019.head()
```

```python
# Only needed data given: So, rename columns to match other years
corp_grade2019 = school_grade2019.rename(columns={"2019 Hold Harmless Grade":"Corp Grade 2019"})
corp_grade2019.head()
```

```python
# Save clean data in Resources directory 
corp_grade2019.to_csv("Resources/schoolgrade2019.csv", index=False, header=True)
```

```python
# Save clean data on other years in Resources directory
corp_grade2015.to_csv("Resources/schoolgrade2015.csv", index=False, header=True)
corp_grade2016.to_csv("Resources/schoolgrade2016.csv", index=False, header=True)
corp_grade2017.to_csv("Resources/schoolgrade2017.csv", index=False, header=True)
corp_grade2018.to_csv("Resources/schoolgrade2018.csv", index=False, header=True)
```
